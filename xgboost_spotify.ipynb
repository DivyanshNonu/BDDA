```python
# 1.0 Call libraries:

import pandas as pd
from  pathlib import Path

# 1.0.1
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

# 1.0.2
from xgboost import XGBRegressor
```

```python
# 1.3 Path:

pathToFolder = "/gdrive/MyDrive/dataset_BDDA"
path = Path(pathToFolder) / "Spotify_Dataset_V2.csv"
```

```python
# 2.0 Read the data
data = pd.read_csv(path,sep=";")
```

```python
# 2.1 Show data:

data.shape   # (13580, 21)
data.head()
```

```python
# 2.2 Select subset of predictors:

cols_to_use = ['Energy', 'Loudness', 'Valence', 'Danceability']
X = data[cols_to_use]

# 2.3 Select target
y = data.Speechiness

# 2.4 Separate data into training and validation sets
X_train, X_valid, y_train, y_valid = train_test_split(X, y)
```

```python
# 3.0
my_model = XGBRegressor()
my_model.fit(X_train, y_train, verbose = True)
```

```python
# 3.1 Predictions:

predictions = my_model.predict(X_valid)
print("Mean Absolute Error: " + str(mean_absolute_error(predictions, y_valid)))
```

```python
# 3.2 Set n_estimators:

my_model = XGBRegressor(n_estimators=500)
my_model.fit(X_train, y_train, verbose = True)
```

```python
# 3.3 Set early stop rounds:

my_model = XGBRegressor(n_estimators=500)
my_model.fit(X_train, y_train,
             early_stopping_rounds=5,
             eval_set=[(X_valid, y_valid)],
             verbose= True)
```

```python
# 3.4 learning_rate:

my_model = XGBRegressor(n_estimators=1000, learning_rate=0.05)
my_model.fit(X_train, y_train,
             early_stopping_rounds=5,
             eval_set=[(X_valid, y_valid)],
             verbose= True)
```
